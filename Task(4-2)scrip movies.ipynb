```python
from selenium import webdriver
import pandas as pd
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
movie_title= driver.find_elements_by_xpath("//td[@class='a-text-left mojo-field-type-title']//a[@class='a-link-normal']")
```

```python
for i in movie_title:
    print(i.text)
```

```python
lifetime=driver.find_elements_by_xpath("//td[@class='a-text-right mojo-field-type-money']")
```

```python
for i in lifetime:
    print(i.text)
```

```python
print(len(movie_title),len(lifetime))
```

```python
movie_title = [element.text for element in movie_title]
lifetime = [element.text for element in lifetime]


df = pd.DataFrame({
    "title": movie_title,
    "lifetime gross": lifetime
})

df.to_csv('movies.csv', index=False, encoding='utf-8')
```
