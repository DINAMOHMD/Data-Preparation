# 2.Exploratory Data Analysis (EDA) and Visualization

###### Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?

```python
num_of_applications =  df.groupby('Name')['Applicants total'].sum().sort_values(ascending = False)
num_of_applications
```

```python
num_of_applications = df[['Name','Applicants total']].sort_values("Applicants total",ascending = False).head(25)
num_of_applications
```

```python
plt.figure(figsize=(20,16))
sns.barplot(x='Applicants total',y='Name',data =num_of_applications)
plt.title('Top 25 Universities with the Most Applications')
plt.xlabel('Number of applications');
```

###### Do students prefer universities that have a high rate of admission?, in other words, do students prefer a university where it is easier for them to be admitted?.


```python
sns.scatterplot(x=df['Applicants total'],y=df['Admissions total'],data=df,hue=df['Control of institution'])
```

```python
sns.scatterplot(x=df['Applicants total'],y=df['Enrolled total'],data=df,hue=df['Control of institution'])
```

```python
sns.scatterplot(x=df['Admissions total'],y=df['Enrolled total'],data=df,hue=df['Control of institution'])
```

```python
df['acceptance_rate'] =(df['Admissions total']/df['Applicants total']* 100) 
df['enrollment_rate'] =(df['Enrolled total']/df['Admissions total']* 100) 
```

```python
sns.scatterplot(x = df.acceptance_rate,y = df.enrollment_rate,data=df,hue=df['Control of institution'])
```

students don't prefer a university where it is easier for them to be admitted
