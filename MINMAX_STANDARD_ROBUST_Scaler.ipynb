```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('moore.csv',header=None)
```

```python
df.columns=['D','X']
```

```python
sns.scatterplot(x=df['D'], y=df['X'])
```

```python
df.head()
```

```python
plt.boxplot(df['X'])
plt.show()
sns.histplot(df['X'], kde=True)
plt.show()
```

# Min Max Scaler

```python
min_max_scaler=(df['X']-df['X'].min())/(df['X'].max()-df['X'].min())
plt.boxplot(min_max_scaler)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data_array = df['X'].to_numpy().reshape(-1, 1)
plt.boxplot(scaler.fit_transform(data_array))
plt.show()
sns.histplot(scaler.fit_transform(data_array), kde=True)
plt.show()
```

```python
print(np.round(np.mean(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))))
print(np.std(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1))))
```

# Z-Score Scaler (standardization)

```python
z_scaler=(df['X']-df['X'].mean())/(df['X'].std())
plt.boxplot(z_scaler)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
plt.boxplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))
plt.show()
sns.histplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)), kde=True)
plt.show()
```

```python
print(np.round(np.mean(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))))
print(np.std(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1))))
```

# Robust Scaler

```python
median = df['X'].median()
iqr = df['X'].quantile(0.75) - df['X'].quantile(0.25)
Robust_Scaler= (df['X'] - median) / iqr
plt.boxplot(Robust_Scaler)
plt.show()
```

```python
from sklearn.preprocessing import  RobustScaler
scaler=RobustScaler()
plt.boxplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))
plt.show()
sns.histplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)), kde=True)
plt.show()
```

```python
print(np.round(np.mean(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))))
print(np.std(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1))))
```

```python
from sklearn.preprocessing import  MaxAbsScaler
scaler=MaxAbsScaler()
plt.boxplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)))
plt.show()
sns.histplot(scaler.fit_transform(df['X'].to_numpy().reshape(-1,1)), kde=True)
plt.show()
```
